```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading data set to only
data = {
    'GEOGRAPHY': ['Central region', 'Eastern region', 'Northern region', 'Western region'],
    'Male': [98330.0, 43787.8, 29029.9, 61765.1],
    'Female': [88538.5, 35201.3, 22445.9, 48838.4]
}

df = pd.DataFrame(data)

plt.figure(figsize=(12, 8))

# Plotting Male unemployment numbers
plt.scatter(df['GEOGRAPHY'], df['Male'], color='blue', label='Unemployed Males', marker='o')

# Plotting Female unemployment numbers
plt.scatter(df['GEOGRAPHY'], df['Female'], color='red', label='Unemployed Females', marker='x')

# Adding labels and title
plt.xlabel('Ontario Regions', fontsize = 18, color = 'Green' )
plt.ylabel('Number of Unemployed Individuals', fontsize = 18, color = 'Green' )
plt.title('Unemployed Males and Females Across Ontario Regions',fontsize = 25, color = 'Green')

for i, row in df.iterrows():
    # Annotate male data
    plt.annotate(f'M: {row["Male"]}', (row['GEOGRAPHY'], row['Male']),
                 textcoords="offset points", xytext=(10, 5), ha='left', color='blue')

    # Annotate female data
    plt.annotate(f'F: {row["Female"]}', (row['GEOGRAPHY'], row['Female']),
                 textcoords="offset points", xytext=(-10, -15), ha='right', color='red')

plt.legend()

plt.show()
```
